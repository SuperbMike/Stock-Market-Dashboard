```python
import requests
import pandas as pd

# Fetch the S&P 500 ticker data from a reliable source
url = 'https://datahub.io/core/s-and-p-500-companies/r/constituents.csv'
response = requests.get(url)

# Read the CSV data into a DataFrame
data = pd.read_csv(url)

# Extract the ticker symbols
sp500_tickers = data['Symbol'].tolist()

# Print the tickers
print(sp500_tickers)
```

```python
to_retrieve = sp500_tickers
to_retrieve
```

```python
import pandas as pd
import numpy as np
import yfinance as yf

data = []

# Fetch and store historical data for each stock
for ticker in to_retrieve:
    try:
        # Fetch historical data using yfinance
        stock_data = yf.download(ticker, start='2000-01-01', end='2023-06-07')
        
        # Create a list of dictionaries for each row of stock data
        rows = [
            {
                'Date': index.strftime('%Y-%m-%d'),
                'Ticker': ticker,
                'Open': row['Open'],
                'High': row['High'],
                'Low': row['Low'],
                'Close': row['Close'],
                'Volume': row['Volume']
            }
            for index, row in stock_data.iterrows()
        ]
        
        data.extend(rows)
        
        print(f"Successfully stored data for {ticker}")
    except Exception as e:
        print(f"Failed to fetch data for {ticker}: {e}")

combined_data = pd.DataFrame.from_records(data)

# Save combined data to a CSV file
combined_data.to_csv("S&P500_price_data.csv", index=False)
```
